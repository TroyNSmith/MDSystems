```python
import matplotlib.pyplot as plt
import numpy as np

from mdsystems import analysis, atoms

test = atoms.System(traj="data/test.xtc", top="data/test.gro")
```

```python
vhs, centers, times = analysis.Dynamics.van_hove_self(test, com=True, rmax=0.5)
```

```python
vhs_plot = np.where(vhs > 1000, np.nan, vhs)
X, Y = np.meshgrid(centers, times)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X, Y, vhs_plot, cmap="inferno", linewidth=0, antialiased=True)

fig.colorbar(surf, shrink=0.5, aspect=8)
ax.set_xlabel("r (nm)")
ax.set_ylabel("t (ps)")
ax.set_zlabel("G_s(r, t)")
ax.set_ylim(0, 100)  # time-range
ax.set_zlim(0, 100)  # G_s range
plt.show()
```
