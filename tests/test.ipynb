```python
import mdtraj as mdt
import numpy as np
from scipy.spatial import cKDTree
```

```python
system = mdt.iterload("data/test.xtc", top="data/test.gro")

for chunk in system:
    frames = []
    for frame in chunk:
        xyz = frame.xyz[0] % frame.unitcell_lengths[0]
        tree = cKDTree(frame.xyz[0] % frame.unitcell_lengths[0], boxsize=frame.unitcell_lengths[0])
        frames.append(tree)
    break
```
